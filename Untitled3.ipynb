```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("F:\R\Random_Forest_Regression\CPSch3 - CPSch3.csv")
data.sex = data.sex.replace(to_replace=['male', 'female'], value=[0, 1])
```

```python
x = data.iloc[10:999,1:3].values
x
```

```python
y = data.iloc[10:999,3].values
y
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=300,random_state=0)
regressor.fit(x,y)
```

```python
var = data.iloc[5:6,1:3]
result = regressor.predict(var)
result
```
